```python
# Load data in grid (dictionary)
grid = {}
with open(r"input.txt", 'r') as file:
    lines = [line for line in file.read().split()]

print(lines)
```

```python
starts = []
for rowIndex, row in enumerate(lines):
    for columnIndex in range(0, len(lines[0])):
        grid[(rowIndex, columnIndex)] = int(row[columnIndex])
        if grid[(rowIndex, columnIndex)]== 0:
            starts.append((rowIndex, columnIndex))

print(rowIndex)
print(columnIndex)
```

```python

def search(x, y, grid, ends):
    min = 0
    max = 55
    height = grid[(x, y)]
    if height == 9: # and (x,y) not in ends:
        ends.append((x,y))
        return
    if x < max and grid[((x+1, y))] == height + 1:
        search(x+1,y, grid, ends)
    if y < max and grid[((x, y+1))] == height + 1:
        search(x,y+1, grid, ends)
    if x > min and grid[((x-1, y))] == height + 1:
        search(x-1,y, grid, ends)
    if y > min and grid[((x, y-1))] == height + 1:
        search(x,y-1, grid, ends)
    return
```

```python
count = 0
ends = [[]]*len(starts)
for startIndex, start in enumerate(starts):
    ends[startIndex] = []
    search(start[0], start[1], grid, ends[startIndex])
    count += len(ends[startIndex])
print(ends)
print(count)
```
